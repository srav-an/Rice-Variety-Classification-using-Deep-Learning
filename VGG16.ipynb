```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import sample

# Machine Learning libraries
from sklearn.model_selection import train_test_split
import tensorflow as tf
import keras
from sklearn.metrics import classification_report

# Libraries for image data
import pathlib
import cv2
```

```python
class ImageDataPrep():
    def __init__(self, dir_path):
        self.dir_path = pathlib.Path(dir_path) # path to the unzipped image data
        self.labels = [pathlib.PurePath(x).name for x in self.dir_path.iterdir() if x.is_dir()] # class labels
        self.label_dict = {x:i for i, x in enumerate(self.labels)} # label dict
        self.image_dict = {l:list(self.dir_path.glob('{}/*'.format(l))) for l in self.labels} # dict for image paths
    
    def randomize_image(self, num_img=100):
        for key, value in dataset.image_dict.items():
            self.image_dict[key] = sample(value, num_img) # sample image subsets
            
    def create_dataset(self, resize=(224,224)):
        self.X, self.y = [], []
        for label, images in self.image_dict.items(): # loop classes
            for image in images: # loop all images for each class
                img = cv2.imread(str(image)) # read image
                resized_img = cv2.resize(img, resize) # resize image for MobNetv2
                self.X.append(resized_img) # add image to X
                self.y.append(self.label_dict[label]) # add label to y
        self.X = np.array(self.X)
        self.X = self.X/255
        self.y = np.array(self.y)
```

```python
def learning_curve(model, X, y, num_pts=10):
    train_sizes, losses, accs = [], [], [] # initialize data lists
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8) # split training/test sets
    batch_size = y_train.shape[0] // num_pts # batch size
    train_size = 0
    for i in range(1,num_pts+1):
        new_model = tf.keras.models.clone_model(model) # refresh model
        train_size += batch_size # number of samples
        if train_size > y_train.shape[0]: # avoid training size exceeding total number of training set
            train_size = y_train.shape[0]
        sub_X, _, sub_y, _ = train_test_split(X_train, y_train, train_size=train_size) # split training/test sets
        # compile model
        new_model.compile(
                  optimizer="adam",
                  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                  metrics=['acc'])
        history = new_model.fit(sub_X, sub_y, epochs=10) # model fit
        loss, acc = new_model.evaluate(X_test, y_test) # model test
        train_sizes.append(train_size)
        losses.append(loss)
        accs.append(acc)
    return train_sizes, losses, accs
```

```python
data_dir = pathlib.Path("/content/Rice_Image_Dataset") # Path to the unzipped image data
dataset = ImageDataPrep(data_dir)
dataset.randomize_image(num_img=804)
dataset.create_dataset()
```

```python
base_model = tf.keras.applications.VGG16(input_shape=(256,256,3), include_top=False, weights='imagenet')
base_model.trainable = False
model = tf.keras.Sequential([
                             base_model,
                   keras.layers.Dense(4096, activation='relu'),
                   keras.layers.Dense(4096, activation='relu'),
                   keras.layers.Dense(5, activation='softmax')])

model.compile(
      optimizer="adam",
      loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
      metrics=['acc'])

model.summary()
```
